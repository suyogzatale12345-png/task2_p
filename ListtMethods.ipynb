### append

```python
color = ['Red','Black','Orange','Pink']
color.append("White")
print(color)
```

### clear

```python
color = ['Red','Black','Orange','Pink']
color.clear()
print(color)
```

### copy

```python
fruits = ["apple", "banana", "cherry"]

x = fruits.copy()

print(x)
```

### count

```python
num = [1,2,5,5,6,7,7,6,10,3]

a = num.count(7)
print(a)
```

### extend Method 

```python
color = ['Red','Black','Orange','Pink']
num = [1,2,5,6]
color.extend(num)
print(color)
```

### index

```python
b = list(range(1,15))
c = b.index(11)
print(c)
```

### insert Method

```python
f = ['Icecreem', 'chocklate', 'cake']

f.insert(1, "water")

print(f)
```

### List POP method

```python
f = ['Icecreem', 'chocklate', 'cake']

z = f.pop(1)

print(z)
```

### remove Method

```python
color = ['Red','Black','Orange','Pink']
color.remove("Red")
print(color)
```

 ### reverse method

```python
movie = ['kgf','pushpa','tere name','animal']
movie.reverse()
print(movie)
```

### sort Method

```python
bike = ['hayabhusa', 'Ninja','goldwing']

bike.sort(reverse=True)

print(bike)
```
